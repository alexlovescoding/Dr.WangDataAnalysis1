```python
from pandas import read_csv

dataframe = read_csv('processed_data.csv')
```

```python
dataframe.tail(100)
```

```python
from numpy import nan
df = dataframe
df['Rlastsec'] = df['Rlastsec'].fillna(0)
df['is_attack'] = nan
df['is_attack'] = (df['Cattemptlastsec'] > 1000) | (df['Rlastsec'] > 3) # This is our rule for our data!

# This is our dataset to train, tweak, and test our model on.
X = df.loc[:, ['Cattemptlastsec', 'Rlastsec']]
Y = df['is_attack'].astype(int)

datalength = len(Y)
x_train = X.loc[:int(datalength*.7)] # however many rows are 70% of the data
y_train = Y.loc[:int(datalength*.7)] # however many rows are 70% of the data

x_cv = X.loc[int(datalength*.7):int(datalength*.9)] # the rows between the 70 % mark and the 90% mark
y_cv = Y.loc[int(datalength*.7):int(datalength*.9)] # the rows between the 70 % mark and the 90% mark

x_test = X.loc[int(datalength*.9):] # however many rows are last 10% of the data
y_test = Y.loc[int(datalength*.9):] # however many rows are last 10% of the data

from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(random_state=0, verbose=1).fit(x_train, y_train)
clf.predict(x_cv[:2])
#clf.predict_proba(X[:2, :])
clf.score(x_train, y_train)
```
